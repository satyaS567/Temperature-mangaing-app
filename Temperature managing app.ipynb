```python
f=open('temps.txt',mode='r')
days=['Monday','Tuesday','Wendsday','Thursday','Friday','Saturaday','Sunday']
temps=[]
for line in f:
    temps.append(float(line[:-1]))
temps_dict=dict(zip(days,temps))
print(temps_dict)
def day_wise():
    for temp in temps_dict:
        print(f"{temp}---->{temps_dict[temp]}")

def average():
    sum1=0
    for temp in temps:
        sum1=sum1+temp
    return sum1/len(days)

def min1():
    min1=temps_dict['Monday']
    day='Monday'
    for temp in temps_dict:
        if temps_dict[temp]<=min1:
            min1=temps_dict[temp]
            day=temp
    return min1,day

def max1():
    max1=temps_dict['Monday']
    for temp in temps_dict:
        if temps_dict[temp]>=max1:
            max1=temps_dict[temp]
            day=temp
    return max1,day

def below_zero():
    print("Below are days having temperature less than or equal zero")
    for temp in temps_dict:
        if temps_dict[temp]<=0:
            print(temp)

def separate_temps():
    sep=float(input("Enter temperature for separation"))
    below_sep=[]
    above_eq_sep=[]
    for temp in temps_dict:
        if temps_dict[temp]<sep:
            below_sep.append(temp)
        else:
            above_eq_sep.append(temp)
    print("Days having temperature below",sep)
    for t in below_sep:
        print(t)
    print("Days having temperature above or equal to",sep)
    for t in above_eq_sep:
        print(t)
while True:
    print("select from following:")
    print("1.Show dayswise temperature\n2.Show average of all temperature\n3.Show minimum temperature\n4.Show maximum temperature\n5.Show days having below zero temperature\n6.Show days using separtor")
    ch=int(input("Enter your choice:"))
    if ch==1:
        day_wise()
    elif ch==2:
        print("Average of temperature for current week is",average())
    elif ch==3:
        minimum=min1()
        print(f"minimum temperature is{minimum[0]}for day{minimum[1]}")
    elif ch==4:
        maximum=max1()
        print(f"minimum temperature is{maximum[0]}for day{maximum[1]}")
    elif ch==5:
        below_zero()
    elif ch==6:
        separate_temps()
    else:
        print("invalid choice")
    ans=input("Do you want to countinue(y/n)").lower()
    if ans!='y':
        break
        
f.close()
```
